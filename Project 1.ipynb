```python
# reviwes = {"Instagram":66577440}

reviews = {}

for i in android:
    app_name = i[0]
    n_reviews = float(i[3])
    # print(n_reviews)
    if app_name not in reviews:
        reviews[app_name] = n_reviews
    elif app_name in reviews and reviews[app_name] < n_reviews:
        reviews[app_name] = n_reviews
```

```python
android_clean = []
already_added = []
for i in android:
    n_reviews = float(i[3])
    app_name = i[0]
    if reviews[app_name] == n_reviews and app_name not in already_added:
        android_clean.append(i)
        already_added.append(app_name)
        
        
```

```python
# Revising a function

def is_english(app_name):
    non_ascii = 0
    for i in app_name:
        if ord(i) > 127:
            non_ascii+=1
            
    if non_ascii>3:
        return False
    else:
        return True
```

# Most Common Genres as per avg installation

```python
avg_installation = []
for cat in uni_cat:
    total_int = 0
    no_ins = 0
    for i in android_final:
        categories = i[1]
        if cat == categories:
            n_install = i[5]
            # print(n_install)
            n_install = n_install.replace("+","").replace(",","")
            n_install = float(n_install)
            total_int += n_install
            no_ins += 1
    avg_ins = total_int/no_ins
    avg_installation.append([cat, avg_ins])
```
